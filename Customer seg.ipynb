```python
x = df[['Age', 'Total Spend']]
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans()
k_means.fit(x)
k_means.fit_predict(x)
```

```python
wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters = i)
    k_means.fit(x)
    wcss.append(k_means.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
k_means = KMeans(n_clusters = 4, random_state = 42)
y_means = k_means.fit_predict(x)
```

```python
y_means
```

```python
plt.scatter(x.iloc[y_means == 0, 0], x.iloc[y_means == 0, 1], s = 100, c = 'red', label = 'Customer 1')
plt.scatter(x.iloc[y_means == 1, 0], x.iloc[y_means == 1, 1], s = 100, c = 'blue', label = 'Customer 2')
plt.scatter(x.iloc[y_means == 2, 0], x.iloc[y_means == 2, 1], s = 100, c = 'green', label = 'Customer 3')
plt.scatter(x.iloc[y_means == 3, 0], x.iloc[y_means == 3, 1], s = 100, c = 'yellow', label = 'Customer 4')
plt.title('Customer Segmentation')
plt.xlabel("Age")
plt.ylabel("Total Spend")
plt.legend()
plt.show()
```
